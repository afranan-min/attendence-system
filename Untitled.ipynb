```python
#as pre!pip install flask flask-mysql-connector

from flask import Flask, render_template_string, request, redirect, url_for, flash
import threading
import webbrowser
import mysql.connector
import subprocess
app = Flask(__name__)
app.secret_key = 'supersecretkey'  # For flash messages

# MySQL connection configuration
db = mysql.connector.connect(
    host="localhost",
    user="root",   # Replace with your MySQL Workbench username
    password="11223344minmin",  # Replace with your MySQL Workbench password
    database="university"  # Replace with your database name
)

# HTML code for the homepage
home_page_html = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Enroll</title>
    <style>
        body {
            background-color: #318f8f; /* Blue background */
            color: white;
            display: flex;
            justify-content: center;
            align-items: center;
            height: 100vh;
            margin: 0;
            font-family: Arial, sans-serif;
        }
        .container {
            text-align: center;
        }
        button {
            background-color: white;
            color:#318f8f;
            border: none;
            padding: 15px 30px;
            margin: 10px;
            font-size: 16px;
            cursor: pointer;
            border-radius: 8px;
        }
        button:hover {
            background-color: #e6e6e6;
        }
    </style>
</head>
<body>
    <div class="container">
        <h1>Choose Enrollment Type</h1>
        <button onclick="window.location.href='/enroll/student'">Enroll as Student</button>
        <button onclick="window.location.href='/start-attendance'">Start Attendance</button>
    </div>
</body>
</html>
"""

# HTML form for student enrollment
student_form_html = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Enroll Student</title>
    <style>
        body {
            background-color: #318f8f;
            color: white;
            display: flex;
            justify-content: center;
            align-items: center;
            height: 100vh;
            font-family: Arial, sans-serif;
            margin: 0;
        }
        .form-container {
            background-color: rgba(255, 255, 255, 0.3);
            padding: 30px;
            border-radius: 10px;
            box-shadow: 0px 0px 15px rgba(0,0,0,0.2);
            text-align: center;
        }
        input, select {
            padding: 10px;
            width: 100%;
            margin: 10px 0;
            border: none;
            border-radius: 5px;
        }
        input[type="submit"] {
            background-color: white;
            color: #318f8f;
            font-weight: bold;
            cursor: pointer;
        }
        input[type="submit"]:hover {
            background-color: #e6e6e6;
        }
        label {
            display: block;
            margin-top: 10px;
            font-weight: bold;
        }
    </style>
</head>
<body>
    <div class="form-container">
        <h2>Student Enrollment</h2>
        <form method="POST" enctype="multipart/form-data">
            <label for="student_id">Student ID</label>
            <input type="text" name="student_id" required>

            <label for="name">Name</label>
            <input type="text" name="name" required>

            <label for="department">Department</label>
            <input type="text" name="department" required>

            <label for="section">Section</label>
            <input type="text" name="section" required>

            <label for="session_year">Session Year</label>
            <input type="text" name="session_year" required>

            <label for="level">Level</label>
            <input type="text" name="level" required>

            <label for="image">Upload Image</label>
            <input type="file" name="image" required>

            <input type="submit" value="Enroll">
        </form>
        {% with messages = get_flashed_messages() %}
          {% if messages %}
            <div>
              {% for message in messages %}
                <p>{{ message }}</p>
              {% endfor %}
            </div>
          {% endif %}
        {% endwith %}
    </div>
</body>
</html>
"""


@app.route('/')
def home():
    return render_template_string(home_page_html)

@app.route('/enroll/student', methods=['GET', 'POST'])
def enroll_student():
    if request.method == 'POST':
        student_id = request.form['student_id']
        name = request.form['name']
        department = request.form['department']
        section = request.form['section']
        session_year = request.form['session_year']
        level = request.form['level']
        image = request.files['image'].read()
        
        cursor = db.cursor()
        sql = "INSERT INTO students (student_id, name, department,section, session_year, level, image) VALUES (%s, %s, %s,%s, %s, %s, %s)"
        val = (student_id, name, department,section, session_year, level, image)
        cursor.execute(sql, val)
        db.commit()
        
        flash("Student enrolled successfully!")
        return redirect(url_for('enroll_student'))
    
    return render_template_string(student_form_html)


    
@app.route('/start-attendance')
def start_attendance():
    # This will start the attendance script
    subprocess.Popen(["jupyter", "nbconvert", "--to", "notebook", "--execute", "--inplace", "attendance_script.ipynb"])
    return "Attendance process started. Close the terminal or press 'q' to stop."
    
def run_app():
    app.run(port=5000)

# Start the Flask app in a separate thread.
thread = threading.Thread(target=run_app)
thread.start()

# Open the web browser to the home page.
webbrowser.open("http://127.0.0.1:5000")
```
